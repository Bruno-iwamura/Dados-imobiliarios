```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# 1. Carrega os dados

df = pd.read_csv('Brasile-real-estate-dataset.csv',encoding='latin-1')

# 2. Remove a coluna vazia ou sem nome

df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# 3. Limpando o nome das colunas

df.columns = [c.strip() for c in df.columns]

# 4. Sobe para o Postgres
engine = create_engine('postgresql://postgres:PostgreeMain@localhost:5432/bruno_db_alura')
df.to_sql('imoveis_brasil', engine, if_exists='replace', index=False)

print("Carga concluída sem erros.")
```
